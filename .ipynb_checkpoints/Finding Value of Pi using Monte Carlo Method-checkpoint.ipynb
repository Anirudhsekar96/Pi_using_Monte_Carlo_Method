```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
inCircle = 0
inSquare = 0
graph = []
for i in range(1,10000000):
    x = np.random.rand()
    y = np.random.rand()
    inSquare = inSquare + 1
    if((x**2 + y**2)<=1):
        inCircle = inCircle + 1
    pi = 4 * inCircle/(inSquare*1.0)
    graph.append(pi)
    
pi = 4 * inCircle/(inSquare*1.0)
```

```python
print(pi)
```

```python
plt.plot(graph)
plt.show()
```

```python
plt.plot(graph)
plt.savefig("pi_converence.png")
```

```python
plt.plot(graph[100000:])
plt.savefig("pi_converence_smaller.png")
```
